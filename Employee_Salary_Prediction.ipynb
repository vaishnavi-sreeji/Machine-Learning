**EMPLOYEE SALARY PREDICTION USING DECSION TREE**

```python
import pandas as pd
df=pd.read_csv("/content/sample_data/salaries.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df["company_new"]=le.fit_transform(df["company"])
df["job_new"]=le.fit_transform(df["job"])
df["degree_new"]=le.fit_transform(df["degree"])
df
```

```python
df.drop(['company','job','degree'],axis=1,inplace=True)
```

```python
df
```

```python
inputs=df.drop("salary_more_then_100k",axis="columns")
target=df[["salary_more_then_100k"]]
inputs
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```
