```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

**SIMPLE LINEAR REGRESSION**

```python
df = pd.read_excel("/content/sample_data/homeprices.csv.xlsx")
print(df)
```

```python
plt.xlabel("price")
plt.ylabel("area")
plt.scatter(df.area,df.price,color="red",marker="+")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["area"]],df.price)
```

```python
reg.predict([[3300]])
```

```python
d = pd.read_excel("/content/sample_data/areas.csv.xlsx")
print(d)
```

```python
p=reg.predict(d)
```

```python
d["price"]=p
print(d)
```

```python
d.to_csv("predictions.csv")
```

```python
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price,color="red",marker="+")
plt.plot(df.area,reg.predict(df[["area"]]),color="blue")
```

**MULTIPLE LINEAR REGRESSION**

```python
df=pd.read_csv("/content/sample_data/homeprices.csv")
df
```

```python
df.bedrooms.median()
```

```python
from math import floor
median_bedrooms=floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms=df.bedrooms.fillna(median_bedrooms)
df.bedrooms
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["area","bedrooms","age"]],df.price)
```

```python
reg.predict([[3000,3,40]])
```
